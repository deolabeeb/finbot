```python
import pandas as pd
```

```python
df = pd.read_csv('data/financial_data.csv')
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue($Million)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income($Million)'].pct_change() * 100
```

```python
df
```

```python
# Fill NA values that result from pct_change calculations with 0 or an appropriate value
df.fillna(0, inplace=True)
```

```python
df
```

## Data Exploration and Analysis

```python
# Average revenue growth for each company
avg_revenue_growth = df.groupby('Company')['Revenue Growth (%)'].mean()
print("Average Revenue Growth (%) by Company:")
print(avg_revenue_growth)
```

**Apple: 1.7%**<br>
Apple’s average revenue growth is positive, indicating a modest increase in revenue over the analyzed period.

<b>Microsoft: 8.3%</b><br>
Microsoft’s average revenue growth shows a significant increase, suggesting strong revenue growth over the period.

<b>Tesla: 23.4%</b><br>
Tesla’s average revenue growth is the highest, reflecting substantial revenue growth over the period.

```python
# Average net income growth for each company
avg_net_income_growth = df.groupby('Company')['Net Income Growth (%)'].mean()
print("\nAverage Net Income Growth (%) by Company:")
print(avg_net_income_growth)
```

<b>Apple: 0.9%</b><br>
Apple's average net income growth is relatively modest, reflecting a slight increase in profitability over the past three years.

<b>Microsoft: 6.1%</b><br>
Microsoft's average net income growth is more substantial, indicating effective profitability growth during this period.

<b>Tesla: 47.3%</b><br>
Tesla’s average net income growth is the highest by a significant margin, showing exceptional improvement in profitability over the past three years.

```python
# Total revenue by company
total_revenue_by_company = df.groupby('Company')['Total Revenue($Million)'].sum()
print("\nTotal Revenue by Company ($Million):")
print(total_revenue_by_company)
```

<b>Apple: $1,143,430 million</b><br>
Apple has the highest total revenue, indicating its significant market presence and substantial revenue generation over the three-year period.

<b>Microsoft: $578,273 million</b><br>
Microsoft’s total revenue is substantial but less than Apple’s, reflecting its strong revenue performance but not as high as Apple’s.

<b>Tesla: $232,058 million</b><br>
Tesla has the lowest total revenue among the three, which is still considerable but lower compared to Apple and Microsoft.
